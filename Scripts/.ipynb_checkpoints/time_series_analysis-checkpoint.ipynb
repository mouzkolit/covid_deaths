```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from statsmodels.tsa.seasonal import seasonal_decompose

data_time = pd.read_csv("/home/physiologie/time_series/data/OGD_gest_kalwo_GEST_KALWOCHE_100.csv", delimiter = ";")
```

```python
# data preprocessing

data_time["Gender"] = [0 if i == "C11-1" else 1 for i in data_time["C-C11-0"]] # encode gender
data_time["Year"] = [i[-6:-2] for i in data_time["C-KALWOCHE-0"]] # encode year
data_time["Kalenderweek"] = [i[-2:] for i in data_time["C-KALWOCHE-0"]] # detect the calender week
data_time["Year_kal"] = [i[-6:] for i in data_time["C-KALWOCHE-0"]] # 
data_time["Date"] = [str(i[-6:-2]) + "-" + str(i[-2:]) for i in data_time["C-KALWOCHE-0"]] # make a year-week-day annotation
data_time["Date"] = [datetime.strptime(i + '-0', "%Y-%W-%w") for i in data_time["Date"]] # format it correctly
```

```python
# bundesländer: 1: Burgenland, 2: Kärnten, 3: Niederösterreich, 4: Oberösterreich,5: Salzburg, 6: Steiermark, 7:
#Tirol, 8: Voralberg, 9: Wien 1: 0-64, 2: 65+
data_time.info()
```

# Exploratory Data Analysis 

```python
# get gender differences
gender = pd.DataFrame(data_time.groupby(["Date","Gender"])["F-ANZ-1"].sum()).reset_index()
gender_male = gender[gender["Gender"] == 0]
gender_female = gender[gender["Gender"] == 1]
gender_male = gender_male.set_index("Date").drop("Gender", axis = 1)
gender_female = gender_female.set_index("Date").drop("Gender",axis = 1)
```

```python
week_death = pd.DataFrame(data_time.groupby(["Kalenderweek", "Gender"])["F-ANZ-1"].sum()).reset_index()
bundesland_death = pd.DataFrame(data_time.groupby(["Kalenderweek", "C-B00-0"])["F-ANZ-1"].sum()).reset_index()

# plot about the Gender distribution
fig,ax = plt.subplots(figsize = (15,5))
sns.despine()
sns.lineplot(x = "Kalenderweek", y = "F-ANZ-1", data = week_death, hue = "Gender")
plt.title("Death Tolls per Gender")

# plot about the Bundesländer distribution
fig,ax = plt.subplots(figsize = (15,5))
sns.despine()
sns.lineplot(x = "Kalenderweek", y = "F-ANZ-1", data = bundesland_death, hue = "C-B00-0")
plt.title("Death Tolls per Bundesland")
```

```python
def bundesland_übersicht(df, choose, gender = None):
    """
    
    Get data for each State in Austria
    
    df--> initial dataframe
    choose --> State choose
    gender --> Compare gender or not
    
    
    """
    
    if gender:
        bezirk = pd.DataFrame(df.groupby(["Date","C-B00-0","Gender"])["F-ANZ-1"].sum()).reset_index()
        bezirk_choosen = bezirk[bezirk["C-B00-0"] == choose]
        bezirk_choosen = bezirk_choosen.set_index("Date").drop("C-B00-0", axis = 1)
        bezirk_choosen_male = bezirk_choosen[bezirk_choosen["Gender"] == 0 ]
        bezirk_choosen_female = bezirk_choosen[bezirk_choosen["Gender"] == 1]
        return bezirk_choosen_male
    
    else:
        bezirk = pd.DataFrame(df.groupby(["Date","C-B00-0"])["F-ANZ-1"].sum()).reset_index()
        bezirk_choosen = bezirk[bezirk["C-B00-0"] == choose]
        bezirk_choosen = bezirk_choosen.set_index("Date").drop("C-B00-0", axis = 1)
        return bezirk_choosen
    

tirol_analysis = bundesland_übersicht(data_time, "B00-7")

bezirks_decompose = seasonal_decompose(tirol_analysis["F-ANZ-1"], period = 52)
z = bezirks_decompose.plot()
```
